```python
import requests
import json

dcardAPI = "https://www.dcard.tw/_api/forums/job/posts?popular=true"
dcardAPIUnPopul = "https://www.dcard.tw/_api/forums/job/posts?popular=false"

resp = requests.get(dcardAPI)
jsnData = json.loads(resp.text)

respUnPopul = requests.get(dcardAPIUnPopul)
jsnDataUnPopul = json.loads(respUnPopul.text)
```

##é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ

```python
print("ä¸€æ¬¡å›å‚³:{0}ç­†è³‡æ–™".format(len(jsnData)))

print("æ•´å€‹json dataç‰©ä»¶å‹æ…‹:", type(jsnData))
print("å–®ç­†json dataç‰©ä»¶å‹æ…‹:", type(jsnData[0]))

print("å–®ç­†è³‡æ–™åŒ…å«:{0}å€‹æ¬„ä½å¦‚ä¸‹:".format(len(jsnData[0].keys())))
for key in jsnData[0].keys():
    print(key)
```

##å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€

```python
jsnData
no = 1
for data in jsnData:
    print("{0}. {1}, {2}, {3}, {4}".format(no,
                                           data["title"],
                                           data["createdAt"],
                                           data["commentCount"], 
                                           data["likeCount"]))
    no += 1
```

##è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€

```python
totalCommentCount = 0
totalLikeCount = 0

for data in jsnData:
    totalCommentCount += data["commentCount"]
    totalLikeCount += data["likeCount"]
    
print("ç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ç‚º:{0}".format(totalCommentCount/len(jsnData))) 
print("ç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ç‚º:{0}".format(totalLikeCount/len(jsnData))) 
   
totalCommentCount = 0
totalLikeCount = 0

for data in jsnDataUnPopul:
    totalCommentCount += data["commentCount"]
    totalLikeCount += data["likeCount"]
    
print("éç†±é–€æ–‡ç« çš„å¹³å‡ç•™è¨€äººæ•¸ç‚º:{0}".format(totalCommentCount/len(jsnData))) 
print("éç†±é–€æ–‡ç« çš„å¹³å‡æŒ‰è®šäººæ•¸ç‚º:{0}".format(totalLikeCount/len(jsnData))) 
    
```
